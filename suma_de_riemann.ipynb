```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import sympify, lambdify, Symbol

def suma_riemann(funcion, variable, intervalo, subintervalos=100):
    # Convertir la función a un objeto simbólico
    x = Symbol(variable)
    f = sympify(funcion)
    f_lambdified = lambdify(x, f, modules=['numpy'])

    # Definir los límites y el ancho de los subintervalos
    a, b = intervalo
    ancho = (b - a) / subintervalos

    # Calcular los puntos de los subintervalos
    x_points = np.linspace(a, b, subintervalos + 1)
    mid_points = (x_points[:-1] + x_points[1:]) / 2  # Puntos medios

    # Calcular la suma de Riemann
    riemann_sum = np.sum(f_lambdified(mid_points) * ancho)
    print(f"Suma de Riemann: {riemann_sum}")

    # Graficar la función y los rectángulos
    x_vals = np.linspace(a, b, 1000)
    y_vals = f_lambdified(x_vals)

    plt.plot(x_vals, y_vals, label=f"f(x) = {funcion}", color='blue')
    plt.bar(mid_points, f_lambdified(mid_points), width=ancho, align='center', alpha=0.3, edgecolor='black', label="Rectángulos")

    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title("Suma de Riemann")
    plt.legend()
    plt.grid(True)
    plt.show()

# Ejemplo de uso
suma_riemann("x**2", "x", (0, 1), 100)
```
